```python
import os
import csv
import json
import torch
import torch.nn as nn
import pandas as pd
import numpy as np
from torch.utils.data import Dataset
from datasets import load_dataset
from tqdm.notebook import tqdm
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments
```

```python
NUM_SAMPLES = 50
OUTPUT_DIR = "/content/processed_poses_embeddings"
TSV_FILE = "/content/metadata_pure.tsv"

os.makedirs(OUTPUT_DIR, exist_ok=True)
if os.path.exists(TSV_FILE):
    os.remove(TSV_FILE)

print(f"Streaming and processing {NUM_SAMPLES} clips from How2Sign")
dataset_hf = load_dataset("bdanko/how2sign-rgb-front-clips", streaming=True)
train_stream = iter(dataset_hf["train"])

with open(TSV_FILE, mode='w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f, delimiter='	')
    writer.writerow(["signal", "output"])

for i in tqdm(range(NUM_SAMPLES), desc="Processing Videos"):
    sample = next(train_stream)
    metadata = json.loads(sample["json"]["bytes"].decode("utf-8")) if isinstance(sample["json"], dict) and "bytes" in sample["json"] else sample["json"]
    ground_truth_text = metadata.get("SENTENCE", metadata.get("sentence", metadata.get("text", "Unknown")))
    video_id = metadata.get("SENTENCE_NAME", f"sample_{i:04d}")
    video_path = f"/tmp/{video_id}.mp4"

    with open(video_path, "wb") as f:
        f.write(sample["mp4"])

    # SAM-3D inference + pooled DINOv3 embeddings
    batch_results, pooled_embeddings = batch_process_video(
        estimator,
        video_path,
        batch_size=16,
        inference_type="full",
        hook_layer_name=HOOK_LAYER_NAME,
    )
    os.remove(video_path)

    sequence_poses = []
    embedding_values = []
    embedding_frame_indices = []

    for frame_idx, frame in enumerate(batch_results):
        if frame and "pred_keypoints_3d" in frame[0]:
            kpts = frame[0]["pred_keypoints_3d"]
            if hasattr(kpts, 'cpu'):
                kpts = kpts.cpu().numpy()
            sequence_poses.append(kpts.squeeze())

            emb = frame[0].get("pooled_dinov3_embedding")
            if emb is None and frame_idx < len(pooled_embeddings):
                emb = pooled_embeddings[frame_idx]
            if emb is not None:
                if hasattr(emb, 'cpu'):
                    emb = emb.cpu().numpy()
                embedding_values.append(np.asarray(emb))
                embedding_frame_indices.append(len(sequence_poses) - 1)

    if sequence_poses:
        payload = {
            "video_id": video_id,
            "ground_truth_text": ground_truth_text,
            "pred_keypoints_3d": np.stack(sequence_poses).astype(np.float32),
            "pooled_dinov3_embedding": {
                "frame_indices": np.asarray(embedding_frame_indices, dtype=np.int32),
                "values": np.stack(embedding_values).astype(np.float32)
                if embedding_values
                else np.empty((0,), dtype=np.float32),
            },
        }

        pose_filepath = os.path.join(OUTPUT_DIR, f"{video_id}.pt")
        torch.save(payload, pose_filepath)

        with open(TSV_FILE, mode='a', newline='', encoding='utf-8') as f:
            writer = csv.writer(f, delimiter='	')
            writer.writerow([os.path.abspath(pose_filepath), ground_truth_text])

```

```python
import torch
import numpy as np
import pandas as pd
from torch.utils.data import Dataset

class PoseDataset(Dataset):
    def __init__(self, tsv_file, tokenizer, max_target_len=128):
        self.data = pd.read_csv(tsv_file, sep='\t')
        self.tokenizer = tokenizer
        self.max_target_len = max_target_len

        # Load the first sample to dynamically determine the combined feature dimension
        sample_payload = torch.load(self.data.iloc[0]['signal'], weights_only=False)

        sample_pose = sample_payload["pred_keypoints_3d"]
        # Calculate flattened pose dim (usually 133 joints * 3 = 399)
        pose_flat_dim = sample_pose.shape[1] * sample_pose.shape[2] if sample_pose.ndim == 3 else sample_pose.shape[1]

        # Calculate DINO dim (if it exists)
        dino_data = sample_payload.get("pooled_dinov3_embedding", {})
        dino_vals = dino_data.get("values", np.array([]))
        dino_dim = dino_vals.shape[1] if len(dino_vals) > 0 else 0

        self.pose_dim = pose_flat_dim + dino_dim

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]

        # 1. Safely load the PyTorch dictionary payload
        payload = torch.load(row['signal'], weights_only=False)

        # 2. Process Poses: Flatten (Frames, Kpts, 3) -> (Frames, Kpts * 3)
        pose_array = payload["pred_keypoints_3d"]
        num_frames = pose_array.shape[0]
        pose_flat = pose_array.reshape(num_frames, -1)

        # 3. Process DINO Embeddings
        dino_data = payload.get("pooled_dinov3_embedding", {})
        dino_vals = dino_data.get("values", np.array([]))
        dino_indices = dino_data.get("frame_indices", np.array([]))
        dino_dim = dino_vals.shape[1] if len(dino_vals) > 0 else 0

        # Create an empty array for DINO features across all frames (handles missing embeddings safely)
        dino_aligned = np.zeros((num_frames, dino_dim), dtype=np.float32)

        if len(dino_indices) > 0:
            # Ensure we only map valid indices to avoid out-of-bounds errors
            valid_mask = dino_indices < num_frames
            valid_indices = dino_indices[valid_mask]
            valid_vals = dino_vals[valid_mask]
            dino_aligned[valid_indices] = valid_vals

        # 4. Concatenate Poses and DINO embeddings along the feature axis
        if dino_dim > 0:
            combined_features = np.concatenate([pose_flat, dino_aligned], axis=-1)
        else:
            combined_features = pose_flat

        pose_tensor = torch.tensor(combined_features, dtype=torch.float32)

        # 5. Tokenize the ground truth text
        labels = self.tokenizer(
            str(row['output']),
            max_length=self.max_target_len,
            truncation=True,
            return_tensors="pt"
        ).input_ids.squeeze()

        return {"pose": pose_tensor, "labels": labels}

def pose_collate_fn(batch):
    poses = [item['pose'] for item in batch]
    labels = [item['labels'] for item in batch]

    max_pose_len = max(p.shape[0] for p in poses)
    pose_dim = poses[0].shape[1]

    padded_poses = torch.zeros(len(poses), max_pose_len, pose_dim)
    attention_mask = torch.zeros(len(poses), max_pose_len, dtype=torch.long)

    for i, p in enumerate(poses):
        length = p.shape[0]
        padded_poses[i, :length, :] = p
        attention_mask[i, :length] = 1

    max_label_len = max(l.shape[0] for l in labels)
    padded_labels = torch.full((len(labels), max_label_len), -100, dtype=torch.long)
    for i, l in enumerate(labels):
        length = l.shape[0]
        padded_labels[i, :length] = l

    return {
        "inputs_embeds": padded_poses,
        "attention_mask": attention_mask,
        "labels": padded_labels
    }
```

```python
import torch.nn as nn

class SpatioTemporalAdapter(nn.Module):
    def __init__(self, input_dim, output_dim, kernel_size=5):
        super().__init__()
        # A kernel_size of 5 means the model looks at the current frame,
        # plus 2 frames backward and 2 frames forward.
        padding = kernel_size // 2

        # We use a two-layer 1D CNN with a ReLU activation in the middle
        self.conv1 = nn.Conv1d(
            in_channels=input_dim,
            out_channels=output_dim,
            kernel_size=kernel_size,
            padding=padding
        )
        self.relu = nn.ReLU()
        self.conv2 = nn.Conv1d(
            in_channels=output_dim,
            out_channels=output_dim,
            kernel_size=kernel_size,
            padding=padding
        )

    def forward(self, x):
        # x comes in as: (Batch, Time, Features)
        # PyTorch Conv1d expects: (Batch, Channels/Features, Time)
        x = x.transpose(1, 2)

        # Pass through the temporal network
        x = self.conv1(x)
        x = self.relu(x)
        x = self.conv2(x)

        # Transpose back to (Batch, Time, Features) for M2M100
        x = x.transpose(1, 2)
        return x

class PoseToTextModel(nn.Module):
    def __init__(self, m2m_model, combined_dim):
        super().__init__()
        self.m2m = m2m_model

        # We replace the simple nn.Linear layer with our Temporal Adapter
        self.adapter = SpatioTemporalAdapter(
            input_dim=combined_dim,
            output_dim=self.m2m.config.d_model,
            kernel_size=5
        )

    def forward(self, inputs_embeds, attention_mask=None, labels=None, **kwargs):
        # inputs_embeds should now be the concatenated [3D_Poses + DINO_Embeddings]
        projected_embeds = self.adapter(inputs_embeds)
        return self.m2m(
            inputs_embeds=projected_embeds,
            attention_mask=attention_mask,
            labels=labels,
            **kwargs
        )

    def generate(self, inputs_embeds, attention_mask=None, **kwargs):
        projected_embeds = self.adapter(inputs_embeds)
        return self.m2m.generate(
            inputs_embeds=projected_embeds,
            attention_mask=attention_mask,
            **kwargs
        )
```

```python
print("\nInitializing Tokenizer and Model...")
tokenizer = M2M100Tokenizer.from_pretrained("facebook/m2m100_418M")
tokenizer.src_lang = "en"
tokenizer.tgt_lang = "en"
base_m2m = M2M100ForConditionalGeneration.from_pretrained("facebook/m2m100_418M")

dataset = PoseDataset(TSV_FILE, tokenizer)
model = PoseToTextModel(base_m2m, dataset.pose_dim)

training_args = Seq2SeqTrainingArguments(
    output_dir="/content/custom_model_output",
    per_device_train_batch_size=1,
    num_train_epochs=150,
    learning_rate=1e-4,
    logging_steps=10,
    save_strategy="no",
    report_to="none",
    remove_unused_columns=False,
    lr_scheduler_type="constant"
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    data_collator=pose_collate_fn,
)

trainer.train()

model.eval()
device = trainer.args.device
model.to(device)

sample_batch = pose_collate_fn([dataset[0]])
with torch.no_grad():
    generated_tokens = model.generate(
        inputs_embeds=sample_batch["inputs_embeds"].to(device),
        attention_mask=sample_batch["attention_mask"].to(device),
        forced_bos_token_id=tokenizer.get_lang_id("en"),
        max_length=50
    )

translated_text = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)
ground_truth = tokenizer.decode(sample_batch["labels"][0][sample_batch["labels"][0] != -100], skip_special_tokens=True)

print(f"\nOriginal Text: {ground_truth}")
print(f"Model Output: {translated_text}")
```

```python
import evaluate
import torch
from tqdm.notebook import tqdm

print("Loading metrics...")
bleu_metric = evaluate.load("bleu")
meteor_metric = evaluate.load("meteor")
chrf_metric = evaluate.load("chrf")
```

```python
model.eval()
device = trainer.args.device
model.to(device)

predictions = []
references = []

print(f"\nRunning inference on all {len(dataset)} training samples...")
with torch.no_grad():
    for i in tqdm(range(len(dataset))):
        sample = dataset[i]
        sample_batch = pose_collate_fn([sample])

        # Generate prediction
        generated_tokens = model.generate(
            inputs_embeds=sample_batch["inputs_embeds"].to(device),
            attention_mask=sample_batch["attention_mask"].to(device),
            forced_bos_token_id=tokenizer.get_lang_id("en"),
            max_length=128
        )

        # Decode prediction
        translated_text = tokenizer.decode(generated_tokens[0], skip_special_tokens=True)

        # Decode ground truth (filtering out the -100 padding tokens)
        label_tokens = sample_batch["labels"][0]
        label_tokens = label_tokens[label_tokens != -100]
        ground_truth = tokenizer.decode(label_tokens, skip_special_tokens=True)

        predictions.append(translated_text)
        references.append(ground_truth)
```

```python
print("\nCalculating final scores...")

bleu_refs = [[ref] for ref in references]
bleu_results = bleu_metric.compute(predictions=predictions, references=bleu_refs)

meteor_results = meteor_metric.compute(predictions=predictions, references=references)
chrf_results = chrf_metric.compute(predictions=predictions, references=references)

print("\n" + "="*40)
print("TRAINING SET MEMORIZATION SCORES")
print("="*40)
# Multiply BLEU and METEOR by 100 to scale them 0-100 to match chrF's default output
print(f"BLEU Score:   {bleu_results['bleu'] * 100:.2f}")
print(f"METEOR Score: {meteor_results['meteor'] * 100:.2f}")
print(f"chrF Score:   {chrf_results['score']:.2f}")
print("="*40)

# Print a few examples to visually inspect the quality
print("\nSample Outputs:")
for i in range(min(3, len(predictions))):
    print(f"\nSample {i+1}:")
    print(f"Truth: {references[i]}")
    print(f"Pred:  {predictions[i]}")
```
